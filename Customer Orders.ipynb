#### 2. Store customer orders:

**PROMPT TEXT:**

* Create a list of 100 customer names
* Store each customer's order details (customer ID, product, price, category) as tuples inside a list
* Use a dictionary where keys are customer names and values are lists of ordered products

**RULES**:
1. **Customer Name** should be in the format f'{last_name}, {first_name}'. Example Mallik, Subhadri
2. **Customer ID** should be a 6 digit integer, unique for each customer. Example, 100123
3. **Unit Price ($)** is the price of 1 unit of the product in USD. Assume all product units as 1 for now.
4. **Category** is text and can be any of: 'Electronics', 'Clothing' and 'Home Essentials'
5. **Product** is text and has to belong to any 1 option chosen in category. Example, Electronics - Mobile Phone, Clothing - Formal Shirt, Home Essentials - Scented Candle. Map upto 10 products for each of the **Category** option  
6. A single customer could have placed multiple orders and a single product might have been ordered by multiple customers as well
7. In the case of multiple orders, store them as a list of tuples against the respective dictionary key
8. There can't be duplicate orders

Following the above, I want you to generate the data synthetically as a dictionary of list of tuples in this notebook 

```python
# Step 1: Generate synthetic customer order data using fixed product prices
import random
from faker import Faker

fake = Faker()

# Define categories and products with fixed prices
categories = {
    'Electronics': [
        'Mobile Phone', 'Laptop', 'Tablet', 'Smart Watch', 'Bluetooth Speaker',
        'Headphones', 'Camera', 'Gaming Console', 'Power Bank', 'Wireless Mouse'
    ],
    'Clothing': [
        'Formal Shirt', 'Jeans', 'Jacket', 'T-Shirt', 'Dress',
        'Sneakers', 'Socks', 'Sweater', 'Shorts', 'Cap'
    ],
    'Home Essentials': [
        'Scented Candle', 'Cookware Set', 'Vacuum Cleaner', 'Bedsheet', 'Pillow',
        'Towel', 'Table Lamp', 'Wall Clock', 'Laundry Basket', 'Water Bottle'
    ]
}

product_prices = {
    'Mobile Phone': 699.00,
    'Laptop': 1299.00,
    'Tablet': 449.00,
    'Smart Watch': 199.00,
    'Bluetooth Speaker': 99.00,
    'Headphones': 149.00,
    'Camera': 799.00,
    'Gaming Console': 499.00,
    'Power Bank': 39.00,
    'Wireless Mouse': 49.00,
    'Formal Shirt': 45.00,
    'Jeans': 65.00,
    'Jacket': 120.00,
    'T-Shirt': 25.00,
    'Dress': 85.00,
    'Sneakers': 95.00,
    'Socks': 12.00,
    'Sweater': 70.00,
    'Shorts': 40.00,
    'Cap': 20.00,
    'Scented Candle': 18.00,
    'Cookware Set': 140.00,
    'Vacuum Cleaner': 220.00,
    'Bedsheet': 75.00,
    'Pillow': 35.00,
    'Towel': 25.00,
    'Table Lamp': 45.00,
    'Wall Clock': 30.00,
    'Laundry Basket': 28.00,
    'Water Bottle': 15.00
}

# Generate 100 unique customer names and IDs
customer_names = set()
customer_ids = set()
while len(customer_names) < 100:
    first = fake.first_name()
    last = fake.last_name()
    name = f'{last}, {first}'
    customer_names.add(name)
customer_names = list(customer_names)

while len(customer_ids) < 100:
    cid = random.randint(100000, 999999)
    customer_ids.add(cid)
customer_ids = list(customer_ids)

# Generate orders for each customer
customer_orders_dict = {}
for i, name in enumerate(customer_names):
    if i < 20:
        num_orders = 5
        preferred_categories = ['Electronics', 'Clothing', 'Home Essentials']
        anchor_product = random.choice(['Laptop', 'Gaming Console'])
        customer_orders = {
            (customer_ids[i], anchor_product, product_prices[anchor_product], 'Electronics')
        }
    elif i < 50:
        num_orders = random.randint(3, 4)
        preferred_categories = ['Electronics', 'Clothing', 'Home Essentials']
        customer_orders = set()
    elif i < 80:
        num_orders = random.randint(2, 3)
        preferred_categories = ['Electronics', 'Clothing', 'Home Essentials']
        customer_orders = set()
    else:
        num_orders = random.randint(1, 3)
        preferred_categories = ['Clothing', 'Home Essentials']
        customer_orders = set()

    while len(customer_orders) < num_orders:
        category = random.choice(preferred_categories)
        if category == 'Electronics' and random.random() < 0.5:
            product = random.choice(['Laptop', 'Gaming Console', 'Smart Watch', 'Camera'])
        else:
            product = random.choice(categories[category])
        price = product_prices[product]
        order_tuple = (customer_ids[i], product, price, category)

        if order_tuple in customer_orders:
            continue
        customer_orders.add(order_tuple)

    customer_orders_dict[name] = list(customer_orders)
```

### 3. Classify products by category

```python
'''
## PROMPT:
* Use a dictionary to map each product to its respective category
* Create a set of unique product categories
* Display all available product categories
'''

import pandas as pd

# Create a table with categories, products and fixed price values
data = []
for category, products in categories.items():
    i = 1
    for product in products:
        price = product_prices[product]
        data.append({
            'Index': i,
            'Category': category,
            'Product': f'{product} (${price:,.0f})'
        })
        i += 1

pdf = pd.DataFrame(data)
pivot_df = pdf.pivot(index='Index', columns='Category', values='Product')
display(pivot_df)
```

### 4. Analyze customer orders

```python
'''
## PROMPT:
* Use a loop to calculate the total amount each customer spends
* If the total purchase value is above 70 percentile, classify the customer as a high-value buyer
* If it is between 40 and 70 percentile, classify the customer as a moderate buyer
* If it is below 40 percentile, classify them as a low-value buyer
* For demonstration, show sample of 5 customers, their purchases, totals and classifications in a table
'''
import pandas as pd

# Calculate total spend per customer using a loop
customer_totals = {}
for customer, orders in customer_orders_dict.items():
    total = 0.0
    for order in orders:
        _, _, price, _ = order
        total += price
    customer_totals[customer] = round(total, 2)

# Determine percentile thresholds
sorted_totals = sorted(customer_totals.values())
count = len(sorted_totals)
threshold_40 = sorted_totals[max(0, int(count * 0.40) - 1)]
threshold_70 = sorted_totals[max(0, int(count * 0.70) - 1)]

# Classify customers by spend
customer_classification = {}
for customer, total in customer_totals.items():
    if total > threshold_70:
        category = 'High-value buyer'
    elif total > threshold_40:
        category = 'Moderate buyer'
    else:
        category = 'Low-value buyer'
    customer_classification[customer] = category

# Build a sample table for 5 customers
sample_customers = list(customer_totals.keys())[:5]
summary_data = []
for customer in sample_customers:
    orders = customer_orders_dict[customer]
    order_details = '; '.join([f"{product} (${price:,.0f})" for _, product, price, _ in orders])
    summary_data.append({
        'Customer': customer,
        'Orders': order_details,
        'Total Spend': f"${customer_totals[customer]:,.0f}",
        'Classification': customer_classification[customer]
    })

summary_df = pd.DataFrame(summary_data)
summary_df
```

### 5. Generate business insights

```python
'''
## PROMPT:
* Calculate the total revenue per product category and store it in a dictionary
* Extract unique products from all orders using a set
* Use a list comprehension to find all customers who purchased "Electronics"
* Identify the top three highest-spending customers using sorting
'''

# Calculate total revenue per product category and store in a dictionary
revenue_by_category = {}
for orders in customer_orders_dict.values():
    for _, _, price, category in orders:
        revenue_by_category[category] = revenue_by_category.get(category, 0.0) + price

# Prepare formatted revenue dictionary for display
formatted_revenue_by_category = {category: f"${total:,.0f}" for category, total in revenue_by_category.items()}

# Extract unique products from all orders using a set
unique_products = {product for orders in customer_orders_dict.values() for _, product, _, _ in orders}

# Find all customers who purchased Electronics using a list comprehension
electronics_customers = [customer for customer, orders in customer_orders_dict.items() if any(category == 'Electronics' for _, _, _, category in orders)]

# Identify the top three highest-spending customers using sorting
top_spenders = sorted(customer_totals.items(), key=lambda item: item[1], reverse=True)[:3]

# Display results
print('Dictionary showing Total revenue by product category:')
print(formatted_revenue_by_category)
print('\nUnique products count:', len(unique_products))
print('\nNumber of customers who purchased Electronics:', len(electronics_customers))
print('\nTop 3 highest-spending customers:')
for customer, total in top_spenders:
    print(f'{customer}: ${total:,.0f}')
```

### 6. Organise and display data

```python
'''
## PROMPT:
* Generate summary of each customer’s total spending grouped by their classification and category, total number of products purchased, 
most expensive product, it's price and customer name
* Use set operations to find customers who purchased from multiple categories; show the 5 highest spenders and the products they purchased
* Identify common customers who bought both electronics and clothing; show the 5 highest spenders among them
'''
import pandas as pd

# Build a flat order table for summary analysis
order_rows = []
for customer, orders in customer_orders_dict.items():
    classification = customer_classification[customer]
    for _, product, price, category in orders:
        order_rows.append({
            'Customer': customer,
            'Classification': classification,
            'Category': category,
            'Product': product,
            'Price': price
        })

order_df = pd.DataFrame(order_rows)

# Group by classification and category, showing count and most expensive product details
idx = order_df.groupby(['Classification', 'Category'])['Price'].idxmax()
most_expensive = order_df.loc[idx, ['Classification', 'Category', 'Product', 'Price', 'Customer']]
most_expensive = most_expensive.rename(columns={
    'Product': 'Most Expensive Product',
    'Price': 'Most Expensive Price',
    'Customer': 'Customer with Most Expensive Purchase'
})

grouped_summary = (
    order_df.groupby(['Classification', 'Category'])
    .agg(Products_Purchased=('Product', 'count'))
    .reset_index()
    .merge(most_expensive, on=['Classification', 'Category'])
)

grouped_summary['Most Expensive Price'] = grouped_summary['Most Expensive Price'].map(lambda x: f'${x:,.0f}')

# Top-5 Tables
order_pivot = order_df.pivot_table(index='Customer', columns='Category', values='Price', aggfunc='sum')
order_pivot['Total Spend'] = order_pivot.sum(axis=1)

mult_orders = order_pivot.copy().dropna()
mult_orders.sort_values('Total Spend', ascending=False, inplace=True)

elec_cloth_only = order_pivot.copy()[['Electronics', 'Clothing']].dropna()
elec_cloth_only['Total Spend'] = elec_cloth_only.sum(axis=1)
elec_cloth_only.sort_values('Total Spend', ascending=False, inplace=True)

print('Summary grouped by Classification and Category:')
display(grouped_summary)
print('\nTop-5 spenders who bought from all categories:')
display(mult_orders.head(5))
print('\nTop-5 spenders who bought only Electronics and Clothing:')
display(elec_cloth_only.head(5))
```
